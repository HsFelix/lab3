```python
import matplotlib.pyplot as plt
import numpy as np

class Componente:
    _frequencia = 2071.4
    _angular = 2*np.pi*_frequencia
    _periodo = 1/_frequencia

    def __init__(self, a, b):
        self.amplitude = a
        self.fase = self._angular*b
    def f(self, t):
        return self.amplitude*np.sin(self._angular*t + self.fase)
    
# Componente(ddp, dt)
Fonte = Componente(3.7, 0)
R = Componente(1.9, -70e-6)
L = Componente(6.4, 5e-6)
C = Componente(2.1, -136e-6)

x = np.linspace(0,Fonte._periodo*1, 150)
plt.figure(dpi=100)
plt.ylabel("Tens√£o (V)")
plt.xlabel("Tempo (s)")
plt.title("Tens√£o nos terminais do circuito RLC")
plt.ticklabel_format(style="sci", axis="x", scilimits=(0,0))
plt.plot(x, Fonte.f(x), label="Fonte")
plt.plot(x, R.f(x), label="R")
plt.plot(x, L.f(x), label="L")
plt.plot(x, C.f(x), label="C")
plt.plot(x, Fonte.f(x)-R.f(x)-L.f(x)-C.f(x), '--k', label="Soma das tens√µes")
plt.legend(loc=0)
plt.show()
```
